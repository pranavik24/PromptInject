```python
from sentence_transformers import SentenceTransformer
import numpy as np

_model = None

def cosine_similarity_embedded(str1: str, str2: str, model_name="all-MiniLM-L6-v2") -> float:
    global _model
    if _model is None:
        _model = SentenceTransformer(model_name)

    v1, v2 = _model.encode([str1, str2])
    dot = np.dot(v1, v2)
    return float(dot / (np.linalg.norm(v1) * np.linalg.norm(v2)))
```
